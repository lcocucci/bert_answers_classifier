```python
import streamlit as st

# Interfaz de Streamlit
st.title('Evaluador de Respuestas Abiertas')

st.write('Ingrese la pregunta, la respuesta correcta y la respuesta del estudiante para obtener un puntaje.')

pregunta = st.text_area('Pregunta')
respuesta_correcta = st.text_area('Respuesta Correcta')
respuesta_estudiante = st.text_area('Respuesta del Estudiante')

if st.button('Evaluar'):
    if pregunta and respuesta_correcta and respuesta_estudiante:
        # Preparar la entrada para el modelo
        inputs = tokenizer(
            text=[respuesta_correcta],
            text_pair=[respuesta_estudiante],
            padding='longest',
            truncation=True,
            return_tensors='pt'
        )
        # Realizar la predicción
        with torch.no_grad():
            outputs = model(**inputs)
            prediccion = outputs.item()
            # Desnormalizar el puntaje
            puntaje = prediccion * 10.0
            # Asegurarse de que el puntaje esté entre 0 y 10
            puntaje = max(0.0, min(10.0, puntaje))
            # Redondear a un decimal
            puntaje = round(puntaje, 1)
            st.success(f'El puntaje asignado es: {puntaje} / 10')
    else:
        st.warning('Por favor, complete todos los campos.')
```

```python
# Ejecutar Streamlit y localtunnel en segundo plano
import threading
import time

# Función para ejecutar Streamlit
def run_streamlit():
    get_ipython().system_raw('streamlit run app.py --server.port 8501 &')

# Función para ejecutar localtunnel y obtener la URL
def run_localtunnel():
    time.sleep(5)  # Esperar a que Streamlit inicie
    print("Iniciando localtunnel...")
    get_ipython().system_raw('lt --port 8501 >> url.txt 2>&1 &')
    time.sleep(5)  # Esperar a que localtunnel genere la URL
    # Leer la URL pública de localtunnel
    with open('url.txt', 'r') as f:
        for line in f:
            if 'https' in line:
                public_url = line.strip()
                print(f'La aplicación está disponible en: {public_url}')
                break
        else:
            print('No se pudo obtener la URL de localtunnel.')

# Ejecutar ambas funciones en hilos separados
thread1 = threading.Thread(target=run_streamlit)
thread2 = threading.Thread(target=run_localtunnel)
thread1.start()
thread2.start()
```
